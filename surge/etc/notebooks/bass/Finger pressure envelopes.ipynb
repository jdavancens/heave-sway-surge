```python
import numpy as np
from surge import *
import matplotlib.pyplot as plt
%matplotlib inline
res = 2000
```

```python
path_stage_3 = Path(
    # 3-1
    BezierCurve((0, 0), (5, 0)),
    BezierCurve((5, 0.1), (9, 0.7)),
    # 3-2
    BezierCurve((9, 0), (20, 0)),
    BezierCurve((20, 0.7), (22, 0.3)),
    # 3-3
    BezierCurve((22, 0), (32, 0)),
    BezierCurve((32, 0.3), (34, 0.5)),
    # 3-4
    BezierCurve((34, 0), (45, 0)),
    BezierCurve((45, 0.5), (47, 0.7)),
)
path_stage_3.set_interpolater(SineInterpolater(freq=20, amp=2))
X = [x/float(res) for x in range(res)]
Y = [path_stage_3(x) for x in X]
fig, ax = plt.subplots(1, 1, figsize=(16, 4))
ax.set_autoscale_on(False)
plt.plot(X,Y, 'r.-')
```
