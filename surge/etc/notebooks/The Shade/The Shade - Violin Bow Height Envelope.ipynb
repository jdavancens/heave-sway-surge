```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
X = [x for x in range(100)]
two_pi = math.pi * 2
```

```python
Y = []
for x in X:
    x = (x/100.) * two_pi
    y = math.sin(2*x) * math.sin(3*x)
    Y.append(y)
plt.plot(X, Y)    
```

```python
Y = []
for x in X:
    x = (x/100.) * two_pi
    y = math.sin(3*x) * math.sin(4*x)
    Y.append(y)
plt.plot(X, Y)  
```

```python
Y = []
for x in X:
    x = (x/100.) * two_pi
    y = math.sin(4*x) * math.sin(5*x)
    Y.append(y)
plt.plot(X, Y)  
```
