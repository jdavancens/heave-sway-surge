```python
from abjad import *
%load_ext abjad.ext.ipython
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
import itertools,random
```

# Represent a ratio as a product of exponentiated prime factors
$$ r=2^p3^q5^r$$

```python
def prime_product(p1,p2,p3):
    return (2**p1) * (3**p2) * (5**p3)
```

```python
powers = range(-2,3)
cps = [cp for cp in itertools.product(powers, repeat=3)]
p = [cp[0] for cp in cps]
q = [cp[1] for cp in cps]
r = [cp[2] for cp in cps]
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(p,q,r,'ro')
ax.set_xlim3d(-2, 2)
ax.set_ylim3d(-2, 2)
ax.set_zlim3d(-2, 2)
```
