    * Fingering Combination, String Selection
* **Physical (Kinematic, Dynamical)**
    * Relative Direction
    * Velocity: rate of change of position with respect to a frame of reference
    * Pressure: force applied perpendicular to the surface of an object per unit area over which that force is distributed
    * Force: cause an object with mass to change its velocity
* **Temporal**
    * Temporal Density, Regularity, Metricity, Pulserate Complexity, Sustain proportion

```python
section_count = 0
measure_count = 1
for i,stage in enumerate(time_signatures):
    for j, section in enumerate(stage):
        tempo = tempo_map[section_count][1].units_per_minute
        print("Stage {}.{}, {} measures, {} bpm".format(i+1, j+1, len(section),tempo))
        for measure in section:
            print("    Measure {}, {}".format(measure_count,str(measure)))
            measure_count += 1
        section_count += 1
```
