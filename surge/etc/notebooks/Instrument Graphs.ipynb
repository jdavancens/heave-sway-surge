```python
import networkx as nx
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
```

```python
inst = ['oboe', 'clarinet', 'saxophone', 'trombone', 'guitar 1', 'guitar 2', 'violin', 'viola', 'cello', 'bass']
mapping = {i:instrument for i, instrument in enumerate(inst)}
n = len(inst)
g = nx.navigable_small_world_graph(n)
g = nx.relabel_nodes(g, mapping)
pos=nx.spring_layout(g)
nx.draw_networkx_nodes(g, pos, alpha=0)
nx.draw_networkx_edges(g, pos)
nx.draw_networkx_labels(g, pos)
plt.axis('off')
```

```python
current_node = choice(inst)
for _ in range(100):
    neighbors = list(g[current_node].keys())
    print(current_node, neighbors)
    current_node = choice(neighbors)
```
