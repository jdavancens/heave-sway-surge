```python
import matplotlib.pyplot as plt
%matplotlib inline
import math
import random
```

```python
y0, y1 = 100, 500
x0, x1 = 0, 1
X = [x / 100. for x in range(100)]
```

```python
def lin(x, y0, y1):
    return y0 + x * (y1 - y0)
```

```python
def quad(x, y0, y1):
    return y0 + x*x * (y1 - y0)
```

```python
def mean(x, y0, y1):
    return y0 + 0.5 * (y1 - y0)
```

```python
def cubic(x, y0, y1):
    return y0 + x**3 * (y1 - y0)
```

```python
def sqrt(x, y0, y1):
    return y0 + math.sqrt(x) * (y1 - y0)
```

```python
def log(x, y0, y1):
    return y0 + math.log(x+1) * (y1 - y0)
```

```python
lin_Y = [lin(x, y0, y1) for x in X]
quad_Y = [quad(x, y0, y1) for x in X]
sqrt_Y = [sqrt(x, y0, y1) for x in X]
```

```python
plt.plot(
    X, lin_Y, 'r', 
    X, quad_Y, 'g', 
    X, sqrt_Y, 'b', 
)
```

```python
class DrunkInterpolater(object):
    def __init__(self, damp):
        self.last_direction = None
        self.damp = damp
    def __call__(self, x, y0, y1):
        if self.last_direction is None:
            direction = random.choice([-1,1])
        else:
            direction = self.last_direction * -1
        self.last_direction = direction
        y = y0 + x * (y1 - y0)
        d_min = abs(y - y0)
        d_max = abs(y1 - y)
        if d_min < d_max:
            d = d_min
        else:
            d = d_max
        dev = random.random() * d
        return y + (direction * dev * self.damp)
```

```python
X = [x / 10. for x in range(10)]
y0, y1 = 0.67, 0.23
drunk = DrunkInterpolater(0.25)
Y = [drunk(x, y0, y1) for x in X]
```

```python
plt.plot(X, Y)
```
