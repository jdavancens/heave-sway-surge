## Put fingering maps in a list sorted by lowest partial

```python
all_fingerings = list()
for k, v in basic_fingering_map.items():
    vector = [[k],v]
    all_fingerings.append(vector)
for k, v in alternate_fingering_map.items():
    vector = [[k],v]
    all_fingerings.append(vector)
for k, v in multiphonic_fingering_map.items():
    vector = [sorted(k),v]
    all_fingerings.append(vector)    
```

```python
all_fingerings = sorted(all_fingerings, key=lambda f: f[0])
all_combis = []
for x in all_fingerings:
    combi = actions.WoodwindFingeringCombination(
        instrument=AltoSaxophone(),
        left=x[1][0],
        right=x[1][1],
        predicted_pitches=x[0]
    )
    all_combis.append(combi)
```

## Show all fingerings that produce either F\#3, F\#4, F\#5 or F\#6 partials (written)

```python
fs_partials = [x for x in all_combis if \
              pitchtools.NamedPitch("fs") in x.predicted_pitches or \
              pitchtools.NamedPitch("fs'") in x.predicted_pitches or \
              pitchtools.NamedPitch("fs''") in x.predicted_pitches or \
              pitchtools.NamedPitch("fs'''") in x.predicted_pitches]
```

```python
chords = []
for fingering_combination in fs_partials: 
    #print(fingering_combination)
    chord = fingering_combination.markup(size=0.67, graphical=False)
    chords.append(chord)
staff = Staff(chords)
show(staff)
```
