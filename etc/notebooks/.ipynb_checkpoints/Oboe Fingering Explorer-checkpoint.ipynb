## Put fingering maps in a list sorted by lowest partial

```python
all_fingerings = list()
```

```python
for k, v in basic_fingering_map.items():
    vector = [[k],v]
    all_fingerings.append(vector)
for k, v in alternate_fingering_map.items():
    vector = [[k],v]
    all_fingerings.append(vector)
for k, v in multiphonic_fingering_map.items():
    vector = [sorted(k),v]
    all_fingerings.append(vector)    
```

```python
all_fingerings = sorted(all_fingerings, key=lambda f: f[0])
all_combis = []
for x in all_fingerings:
    combi = actions.WoodwindFingeringCombination(
        instrument=oboe,
        left=x[1][0],
        right=x[1][1],
        predicted_pitches=x[0]
    )
    all_combis.append(combi)
```

# Random string of related fingerings

```python
s = Staff("c'2 c'4. c'8 r8 c'4 c'8 c'2")
t = Staff("c'8 c'4. c'2 c'4 c'4 ~ c'8 c'4. ")
score = Score([s,t])
show(score)
```

```python
def pick_new_fingering(simult, last_left, last_right, left_set, right_set, combis):
    new_fingering = None
    history = []
    # pick new random fingering combination and test if in the master set
    while (new_fingering is None or new_fingering not in combis) and len(history) < len(combis):
        if simult is -1:
            new_left = random.choice(left_set)
            while last_left == new_left:
                new_left = random.choice(left_set)
            new_fingering = actions.WoodwindFingeringCombination(
                instrument=oboe,
                left=new_left,
                right=last_right,
            )
        elif simult is 0:
            new_left = random.choice(left_set)
            new_right = random.choice(right_set)
            while (last_left == new_left) and (last_right == new_right):
                new_left = random.choice(left_set)
                new_right = random.choice(right_set)
            new_fingering = actions.WoodwindFingeringCombination(
                instrument=oboe,
                left=new_left,
                right=new_right,
            )
        else:
            new_right = random.choice(right_set)
            while last_right == new_right:
                new_right = random.choice(right_set)
            new_fingering = actions.WoodwindFingeringCombination(
                instrument=oboe,
                left=last_left,
                right=new_right,
            )
        if new_fingering not in history:
            history.append(new_fingering)
    # if none could be found make any random fingering
    if new_fingering is None:
        if simult is -1:
            new_left = random.choice(left_set)
            new_fingering = actions.WoodwindFingeringCombination(
                instrument=oboe,
                left=new_left,
                right=last_right,
            )
        elif simult is 0:
            new_left = random.choice(left_set)
            new_right = random.choice(right_set)
            new_fingering = actions.WoodwindFingeringCombination(
                instrument=oboe,
                left=new_left,
                right=new_right,
            )
        else:
            new_right = random.choice(right_set)
            new_fingering = actions.WoodwindFingeringCombination(
                instrument=oboe,
                left=last_left,
                right=new_right,
            )
        
    return new_fingering
```

```python
# split left and right hand fingerings from set
# randomly choose a new (different) one from set, if none that are different, pick any
left = list(set([f.left for f in my_partials]))
right = list(set([f.right for f in my_partials]))    
```

```python
last_left = actions.WoodwindFingering(
    instrument=oboe, 
    hand='left',
    keys={
        'thumb':None,
        'index':None,
        'middle':None,
        'ring':None,
        'pinky':None
    }
)
last_right = actions.WoodwindFingering(
    instrument=oboe, 
    hand='right',
    keys={
        'index':None,
        'middle':None,
        'ring':None,
        'pinky':None
    }
)
```

```python
fingerings = []
for vertical_moment in iterate(score).by_vertical_moment():
    offset = vertical_moment.offset
    s_notes = (inspect(s).get_vertical_moment_at(offset).start_notes)
    t_notes = (inspect(t).get_vertical_moment_at(offset).start_notes)
    if (s_notes and not t_notes):
        simult = -1 
    elif (s_notes and t_notes):
        simult = 0
    elif (not s_notes and t_notes):
        simult = 1
    new_fingering = pick_new_fingering(simult, last_left, last_right, left, right, my_partials)
    fingerings.append(new_fingering)    
    last_left = new_fingering.left
    last_right = new_fingering.right
```

```python
my_fingerings = []
for f in fingerings:
    i = my_partials.index(f)
    my_fingerings.append(my_partials[i])
```

```python
chords = []
for fingering_combination in my_fingerings: 
    chord = fingering_combination.markup(size=0.67, graphical=False)
    chords.append(chord)
staff = Staff(chords)
show(staff)
```

```python
def pitch_filter(pitches, fingerings):
    pitches = [pitchtools.NamedPitch(pitch) for pitch in pitches]
    filtered_fingerings = []
    for fingering in fingerings:
        for pitch in pitches:
            if pitch in fingering.predicted_pitches:
                filtered_fingerings.append(fingering)
    return filtered_fingerings
```
