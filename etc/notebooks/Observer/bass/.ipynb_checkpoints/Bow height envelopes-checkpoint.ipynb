```python
import numpy as np

from surge.tools.rhythmtools.sum_time_signatures import sum_time_signatures
from surge.tools.rhythmtools.time_signatures_to_offsets import time_signatures_to_offsets
import matplotlib.pyplot as plt
%matplotlib inline
res = 2000
from surge.materials.segment_i.time_signatures import time_signatures
```

```python
offsets_stage_3 = time_signatures_to_offsets(time_signatures[2])
total_duration = float(sum_time_signatures(time_signatures[2]))

path_stage_3 = surge.tools.bezzie.Path(
    BezierCurve((offsets_stage_3[0], 1), (total_duration, 0.8)),
)
path_stage_3.set_interpolater(SineInterpolater(freq=120, amp=2))
X = [x for x in offsets_stage_3]
Y = [path_stage_3(x) for x in offsets_stage_3]
fig, ax = plt.subplots(1, 1, figsize=(16, 4))
ax.set_autoscale_on(True)
plt.plot(X,Y, 'r-')
```
