```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt

res = 2000
%matplotlib inline
```

```python
path_stage_1 = surge.bezzie.Path(
    # 1-1
    BezierCurve((0, 0.2), (16, 0.5)),
    # 1-2
    BezierCurve((16, 0.3), (36, 0.7)),
)
path_stage_1.set_interpolater(SineInterpolater(freq=80, amp=2.3))
X = [x/float(res) for x in range(res)]
Y = [path_stage_1(x) for x in X]
fig, ax = plt.subplots(1, 1, figsize=(16, 4))
ax.set_autoscale_on(False)
plt.plot(X,Y, 'r.-')
```
